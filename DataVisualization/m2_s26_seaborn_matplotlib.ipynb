```python
import pandas as pd

# Sample data
data = {
    "Month": [
        "Jan","Jan","Jan","Jan",
        "Feb","Feb","Feb","Feb",
        "Mar","Mar","Mar","Mar"
    ],
    "Region": [
        "North","South","East","West",
        "North","South","East","West",
        "North","South","East","West"
    ],
    "Sales": [
        12000, 10000, 9000, 11000,
        15000, 11000, 9500, 12000,
        13000, 11500, 10000, 12500
    ]
}

df = pd.DataFrame(data)

print(df)
```

```python
import matplotlib.pyplot as plt

import seaborn as sns


# Set up figure with subplots (1 row, 3 cols)

fig, axes = plt.subplots(1, 3, figsize=(20, 5))


# 1. Line Plot: Monthly Sales

sns.lineplot(data=df, x='Month', y='Sales', ax=axes[0], ci=False)

axes[0].set_title('Monthly Trend')


# 2. Bar Chart: Regional Sales

sns.barplot(data=df, x='Region', y='Sales', ax=axes[1])


# TODO: Add Histogram on axes[2] and show plot
sns.histplot(data=df, x='Sales', ax=axes[2])
plt.show()
```
